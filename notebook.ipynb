## Conclusion


### Bagaimana tren peningkatan jumlah pengguna rental bike setiap bulannya?

```python
monthly_rentalbike_df = day_hour_df.groupby(by="mnth").agg({
    "cnt" : "nunique"
})
monthly_rentalbike_df.index = monthly_rentalbike_df.index.map({
    1: 'January', 2: 'February', 3: 'March', 4: 'April', 
    5: 'May', 6: 'June', 7: 'July', 8: 'August', 
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
})

plt.figure(figsize=(10, 5)) 
plt.plot(monthly_rentalbike_df["cnt"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Number of User Rental Bike (2011)", loc="center", fontsize=20) 
plt.xticks(fontsize=10, rotation=25) 
plt.yticks(fontsize=10) 
plt.show()
```

Dari hasil visualisasi yang diperoleh, dapat disimpulkan bahwa terjadi peningkatan signifikan dalam jumlah pengguna rental bike sejak awal tahun. Jumlah pengguna mencapai titik tertinggi antara bulan Mei hingga November, menunjukkan periode puncak penggunaan. Setelah mencapai puncaknya, terjadi penurunan jumlah pengguna pada bulan-bulan akhir tahun, yang menunjukkan adanya penurunan minat atau penggunaan rental bike saat memasuki musim dingin atau akhir tahun.

### Pada hari-hari apa saja masyarakat sering menggunakan rental bike?

```python
day_rentalbike_df = day_hour_df.groupby(by="weekday").agg({
    "cnt" : "nunique",
})
day_rentalbike_df.index = day_rentalbike_df.index.map({
    0:'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5:'Friday', 6:'Saturday'
})

plt.figure(figsize=(10, 5)) 
plt.plot(day_rentalbike_df["cnt"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Number of User Rental Bike (Day)", loc="center", fontsize=20) 
plt.xticks(fontsize=10, rotation=25) 
plt.yticks(fontsize=10) 
plt.show()
```

Dari grafik yang ditampilkan, terlihat bahwa penggunaan rental bike mencapai titik tertinggi pada hari Rabu, menunjukkan bahwa hari tersebut adalah puncak aktivitas penggunaan. Sebaliknya, penggunaan rental bike berada pada titik terendah pada hari Minggu, yang mengindikasikan bahwa aktivitas penggunaan cenderung menurun pada akhir pekan. Pola ini mungkin mencerminkan kebiasaan masyarakat yang lebih aktif pada pertengahan minggu dibandingkan dengan akhir pekan, atau mungkin adanya faktor lain yang mempengaruhi preferensi penggunaan rental bike di hari-hari tertentu.
